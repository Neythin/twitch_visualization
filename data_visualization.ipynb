```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
twitch = pd.read_csv("peak_view_streamers.csv")
twitch
```

# Data Visualization

```python
plt.scatter(twitch["Followers"], twitch["Average Viewers"])
plt.xlabel("Followers (in ten millions)")
plt.ylabel("Average Viewers")
plt.title("Average Viewer vs Follower Count")
plt.xlim(0, 16000000)
plt.ylim(0, 230000)
```

```python
plt.hist(twitch['Peak Viewers'])
```

```python
plt.bar(twitch["Region"], twitch["Followers"])
plt.xlabel("Region")
plt.ylabel("Followers (in hundred of millions)")
plt.title("Cumulative Followers by Region")
```

```python
plt.bar(twitch["Region"], twitch["Average Viewers"])
plt.xlabel("Region")
plt.ylabel("Average Viewers")
```

```python
plt.bar(twitch["Region"], twitch["Peak Viewers"])
plt.xlabel("Region")
plt.ylabel("Peak Viewership")
```

```python
region_proportions = twitch.groupby("Region").size() / len(twitch)
```

```python
plt.pie(region_proportions, labels = twitch["Region"].unique())
plt.legend(title = "Region")
plt.show()
```
