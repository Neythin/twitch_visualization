```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
import seaborn as sns
from collections import Counter
```

```python
twitch = pd.read_csv("peak_view_streamers.csv")
twitch['Popular Categories'] = twitch['Popular Categories'].apply(ast.literal_eval)
twitch
```

# Data Visualization

## Scatterplot

```python
plt.scatter(twitch["Followers"], twitch["Average Viewers"])
plt.title("Average Viewer vs Follower Count")
plt.xlabel("Followers (in Tens of Millions)")
plt.ylabel("Average Viewers")
plt.show()
```

```python
plt.scatter(twitch["Followers"], twitch["Peak Viewers"])
plt.show()
```

```python
plt.scatter(twitch["Peak Viewers"], twitch["Average Viewers"])
plt.show()
```

## Histogram

```python
plt.hist(twitch['Peak Viewers'])
plt.show()
```

```python
plt.hist(twitch['Average Viewers'])
plt.show()
```

```python
plt.hist(twitch['Followers'])
plt.show()
```

## Bar Graph

```python
plt.bar(twitch["Region"], twitch["Followers"])
plt.title("Cumulative Followers by Region")
plt.xlabel("Region")
plt.ylabel("Followers (in hundred of millions)")
plt.show()
```

```python
flat_list = [title for sublist in twitch["Popular Categories"] for title in sublist]

title_counts = Counter(flat_list)
title_counts_values = list(title_counts.values())
title_counts_keys = list(title_counts.keys())

plt.barh(title_counts_keys, title_counts_values)
plt.show()
```

```python
flat_platform = [platform for platform in twitch["Platform"]]

platform_counts = Counter(flat_platform)
platform_values = list(platform_counts.values())
platform_keys = list(platform_counts.keys())

plt.bar(platform_keys, platform_values)
plt.show()
```

## Pie Chart

```python
region_proportions = twitch.groupby("Region").size() / len(twitch)

plt.pie(region_proportions, labels = twitch["Region"].unique())
plt.legend(title = "Region")
plt.show()
```

## Box and Whisker Plot

```python
sns.boxplot(x="Region", y="Average Viewers", data=twitch)
plt.show()
```
